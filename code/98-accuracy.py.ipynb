```python
path = "../input/sms-spam-collection-dataset/spam.csv"
df = pd.read_csv(path, on_bad_lines="skip", encoding="ISO-8859-1")
df = df.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
df.columns = ["label", "sms"]
df.head()
```

```python
import re

corpus = []

for i in range(len(df)):
    review = re.sub("[^a-zA-Z]", " ", df["sms"][i])
    review = review.lower()
    review = review.split()
    review = ' '.join(review)
    corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

cv = CountVectorizer(max_features=2500)
X = cv.fit_transform(corpus).toarray()
y = pd.get_dummies(df["label"], drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X, y.values.ravel(), test_size=0.20, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
clf_lr = LogisticRegression(solver="liblinear", C=10, penalty="l2")
clf_lr.fit(X_train, y_train)
pred_lr = clf_lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred_lr))
```
